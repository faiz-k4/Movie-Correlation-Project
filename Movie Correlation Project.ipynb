```python
# importing libraries

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# reading the dataset

df = pd.read_csv(r'C:\Users\ASUS\Downloads\movies.csv')
```

```python
# take a look at the datset

df.head()
```

```python
df.describe
```

```python
# checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# removing null values

df1 = df.dropna()
```

```python
# describing new dataset after removing null values

df1.describe
```

```python
# checking for missing data again

for col in df1.columns:
    pct_missing = np.mean(df1[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df = df1
```

```python
# Data type of our columns

df.dtypes
```

```python
# change data type of columns

df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
df.dtypes
```

```python
df = df.sort_values(by=['gross'], inplace = False, ascending = False)
df
```

```python
#Setting the screen to fetch and shows all the values in data

pd.set_option('display.max_rows', None)
```

```python
# Drop duplicates

df.drop_duplicates()
```

```python
# Setting the screen to fetch and shows only 20 values of data as its was taking t o fetch
# delete this cell when needed

pd.set_option('display.max_rows', 20)
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earning')

plt.show
```

```python
# plot budget vs gross using Seaborn

sns.regplot(x='budget', y='gross', data = df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Finding Correlation Coefficient

correlation_matrix = df.corr(numeric_only=True)
correlation_matrix
```

```python
# Plotting heatmap of correlation matrix

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation Matrix for Numeric values")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")

plt.show()
```

```python
# to find correlation of all columns
# converted the column to numeric columns

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
df
```

```python
#finding Correlation with all the columns

correlation_matrix1 = df_numerized.corr(numeric_only=True)
correlation_matrix1

sns.heatmap(correlation_matrix1, annot = True)

plt.title("Correlation Matrix for Numeric values")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")

plt.show()
```

```python
#changing the color to make it more understandable

correlation_matrix1 = df_numerized.corr(numeric_only=True)
correlation_matrix1

sns.heatmap(correlation_matrix1, annot = True, cmap = "Blues")

plt.title("Correlation Matrix for Numeric values")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# Filtering high correlation

high_corr = sorted_pairs[(sorted_pairs)> 0.5]
high_corr
```
